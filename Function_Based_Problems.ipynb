```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_post_engagement(likes):
    total_likes = 0
    for like in likes:
        total_likes += like
    
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    print("Total Likes:", total_likes)
    print("Post Status:", status)

analyze_post_engagement([200, 300, 250, 300])
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System
def medicine_stock_alert(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"
    print("Medicine Stock:", stock)
    print("Status:", status)

medicine_stock_alert(6)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain
    average = total / len(rainfall_data)
    
    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall_checker([70, 80, 65, 75], 70)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

detect_duplicates(["user1", "user2", "user3", "user1"])
```

```python
# Problem 5: Healthcare – Appointment Eligibility Checker
def check_eligibility(age):
    status = "Eligible" if age >= 18 else "Not Eligible"
    print("Patient Age:", age)
    print("Eligibility Status:", status)

check_eligibility(21)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
def filter_premium_crops(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)
    print("Premium Crops:", premium)

filter_premium_crops([1500, 2500, 1800, 3200])
```

```python
# Problem 7: System Monitoring – Application Health Checker
def application_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", errors)
    print("System Status:", status)

application_health(7)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
def transaction_checker(amount):
    limit = 50000
    status = "Approved" if amount <= limit else "Rejected"
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_checker(60000)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System
def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = 0
    for status in attendance_list:
        if status == 1:
            attended += 1
    
    percentage = (attended / total_classes) * 100
    eligibility = "Eligible" if percentage >= 75 else "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)

attendance_checker([1,1,1,0,1])
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def electricity_bill(units):
    bill = 0
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    return bill, usage

bill_amount, usage_status = electricity_bill(250)
print("Total Bill:", bill_amount)
print("Usage Classification:", usage_status)
```
