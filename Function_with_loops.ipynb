### Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False

    if amount not in valid_plans:
        print("Invalid plan! Available plans:", valid_plans)
        return False

    return True


# Main Program
while True:
    recharge_amount = int(input("Enter recharge amount: ₹"))

    if validate_recharge(recharge_amount):
        print("Recharge successful!")
        break
    else:
        print("Please try again...\n")
```

### Inventory Reorder Alert System

```python
def check_stock(products):
    for item, quantity in products.items():
        if quantity < 15:
            print(f"{item}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{item}: Stock OK (Stock = {quantity})")


# Sample Inventory
inventory = {
    "Rice": 50,
    "Sugar": 12,
    "Milk": 8,
    "Oil": 25,
    "Salt": 15
}

# Function Call
check_stock(inventory)
```

### Student Result Processing System

```python
def evaluate_result(marks):
    total = 0
    
    # calculate total using loop
    for m in marks:
        total += m
    
    average = total / len(marks)

    # return result
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


# Example student marks
student_marks = [65, 40, 55, 70, 45]

avg, result = evaluate_result(student_marks)

print("Average Marks:", avg)
print("Result:", result)
```

### Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    # apply peak hour charge
    if peak.lower() == 'y':
        fare += fare * 0.25

    return fare


# Main Program with retry
while True:
    distance = float(input("Enter distance travelled (km): "))
    peak = input("Is it peak hour? (y/n): ")

    if distance <= 0:
        print("Distance must be greater than 0\n")
        continue

    total_fare = calculate_fare(distance, peak)

    print(f"Total Fare: ₹{total_fare:.2f}")
    break
```

### Employee Attendance Eligibility Checker

```python
def check_attendance(attendance):
    present_days = 0

    # count present days using loop
    for day in attendance:
        if day.upper() == "P":
            present_days += 1

    total_days = len(attendance)
    percentage = (present_days / total_days) * 100

    # return eligibility
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Example attendance record
record = ["P", "A", "P", "P", "A", "P", "P", "P"]

percent, status = check_attendance(record)

print("Attendance Percentage:", percent)
print("Status:", status)
```

### Password Strength Checker

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # condition 1: minimum length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    # loop through string
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    # final validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"


# Example
pwd = input("Enter password: ")
result = check_password_strength(pwd)
print(result)
```
