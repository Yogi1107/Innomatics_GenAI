# Problem 1: Smart Parking Lot Management System

Design a function to manage a smart parking lot system.
- Accept vehicle entry and exit logs
- Calculate total parked vehicles
- Identify peak parking usage
- Alert if parking exceeds capacity

```
def smart_parking(capacity, logs):
    current = 0
    peak = 0
    
    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT" and current > 0:
            current -= 1
        
        peak = max(peak, current)
    
    status = "Available" if current <= capacity else "Full"
    
    print("Currently Parked Vehicles:", current)
    print("Peak Usage:", peak)
    print("Parking Status:", status)

# Sample Input
smart_parking(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

# Problem 2: Online Food Delivery Time Estimator

Estimate delivery time based on distance, traffic, and weather conditions.

```
def delivery_time(distance, traffic, weather):
    base_time = distance * 5
    
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 8
    
    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20
    
    print("Estimated Delivery Time:", base_time, "minutes")

# Sample Input
delivery_time(8, "High", "Rainy")
```

# Problem 3: Movie Theatre Seat Occupancy Analyzer

Analyze seat booking data to determine occupancy and show status.

```
def theatre_occupancy(total_seats, booked_list):
    booked = len(booked_list)
    occupancy = (booked / total_seats) * 100
    
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"
    
    print(f"Occupancy: {occupancy:.0f}%")
    print("Show Status:", status)

# Sample Input
theatre_occupancy(200, [1]*150)
```

# Problem 4: Cloud Server Load Classification System

Classify server load based on average CPU usage.

```
def server_load(cpu_readings):
    avg = sum(cpu_readings)/len(cpu_readings)
    
    if avg < 50:
        status = "Normal"
    elif 50 <= avg <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    print(f"Average CPU Load: {avg:.0f}%")
    print("Server Status:", status)

# Sample Input
server_load([45, 60, 70, 85, 90])
```

# Problem 5: Smart Classroom Resource Usage Monitor

Track usage of classroom resources and identify overuse.

```
def resource_monitor(resources):
    overused = [r for r, h in resources.items() if h > 8]
    
    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", "Yes" if overused else "No")

# Sample Input
resource_monitor({
 "Projector": 6,
 "AC": 9,
 "Lights": 4
})
```

# Problem 6: Online Event Registration Capacity Controller

Manage registrations and trigger waitlist if capacity exceeded.

```
def event_registration(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlisted = max(0, registrations - capacity)
    status = "Closed" if registrations >= capacity else "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

# Sample Input
event_registration(100, 105)
```
