### Employee Performance Bonus Eligibility

```python
# Dictionary containing employee names and their respective performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Identify the highest score achieved among all employees
max_score = max(employees.values())
print("Top Performers eligible for Bonus:-", end=" ")
# Iterate through the dictionary to find all employees matching the maximum score
for i in employees.keys():
    if employees[i] == max_score:
        # Print names on the same line, separated by a space
        print(f"{i}", end=" ")
```

### Search Query Keyword Analysis

```python
sentence = "Buy mobile phone buy phone online"
# Standardize text to lowercase and split into individual words
sentence_list = sentence.lower().split()
d = {}
result = {}
# Create a frequency map of every word in the sentence
for i in sentence_list:
    d[i] = sentence_list.count(i)
# Filter the map to keep only words that appear more than once
for i in d.keys():
    if d[i] > 1:
        result[i] = d[i]
print(result)
```

### Sensor Data Validation

```python
# List of raw data from the sensors
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
result = []
# Loop through readings while keeping track of their position (index)
for index, value in enumerate(sensor_readings):
    # Filter for even numbers only
    if value % 2 == 0:
        # Store the position and the value as a pair (tuple)
        t = (index, value)
        result.append(t)
# Output the list of found pairs
print(result)
```

### Email Domain Usage Analysis

```python
# List of user emails to be processed
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Extract only the domain names from the email addresses
domains = []
for i in emails:
    domains.append(i.split('@')[1])

# Count the frequency of each domain
domain_counts = {}
for i in domains:
    domain_counts[i] = domains.count(i)

# Calculate and display the percentage share of each domain
for domain in domain_counts.keys():
    percent = domain_counts[domain] / len(domains) * 100
    print(f"{domain}:{int(percent)}%")
```

### Sales Spike Detection

```python
# Input sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]
# Calculate the average daily sales
avg_sales = sum(sales) / len(sales)
# Multiplying by 1.3 is the same as avg + (avg * 0.30)
threshold = avg_sales * 1.3
# Detect and display days that exceed the threshold
for index, value in enumerate(sales):
    if value > threshold:
        # We use index + 1 because day numbering usually starts at 1, not 0
        print(f"Day {index + 1}: {value}")
```

### Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
d = {}
for i in user_ids:
    d[i] = user_ids.count(i)
# Checking if there are any duplicates using the frequency
# If frequency is greater than 1 then its a duplicate
for i in d.keys():
    if d[i] > 1:
        print(f"{i} -> {d[i]} times")
```
