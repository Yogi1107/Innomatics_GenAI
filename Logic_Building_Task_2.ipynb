### Problem 1: Unique Words in a Sentence
#### Identify all unique words and display their count.

```python
sentence = "python is easy and python is powerful"

# Split the sentence into a list of words
words = sentence.split()

# Convert list to a set to filter out duplicates
unique_words = set(words)

print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

### Problem 2: Highest Salary from Employee Data
#### Determine which employee has the highest salary from a dictionary.

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find the key with the maximum value
highest_paid_emp = max(employees, key=employees.get)
highest_salary = employees[highest_paid_emp]

print(f"Highest Salary: {highest_paid_emp} - {highest_salary}")
```

### Problem 3: Find Maximum and Minimum Values
#### Find the max and min values in a list using a loop.

```python
numbers = [45, 22, 89, 10, 66]

# Initializing with the first element
max_val = numbers[0]
min_val = numbers[0]

# Iterating through the list to compare values
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print(f"List: {numbers}")
print(f"Max: {max_val}")
print(f"Min: {min_val}")
```

### Problem 4: Count Products Above a Price Threshold
#### Count products with a price greater than 1000.

```python
# %%
prices = [450, 1200, 899, 1500, 300]
threshold = 1000
count = 0

for price in prices:
    if price > threshold:
        count += 1

print(f"Products above 1000: {count}")
```

### Problem 5: Calculate Attendance Percentage
#### Calculate percentage based on "P" for Present and "A" for Absent.

```python
# %%
attendance = ["P", "P", "A", "P", "P"]

# Count occurrences of 'P'
present_days = attendance.count("P")
total_days = len(attendance)

# Formula: (Part / Total) * 100
percentage = (present_days / total_days) * 100

print(f"Attendance Percentage: {percentage}")
```

# ### Problem 7: Count Character Frequency
# Count how many times each character appears in a string.

```python
# %%
text = "pythonp"
frequency = {}

for char in text:
    # If char is already in dict, increment; else, set to 1
    frequency[char] = frequency.get(char, 0) + 1

print(frequency)
```

# ### Problem 9: Check if a Key Exists in a Dictionary
# Verify the presence of a key using the 'in' keyword.

```python
inventory = {"Apples": 50, "Bananas": 30, "Oranges": 20}
search_key = "Apples"

if search_key in inventory:
    print("Employee exists") # Matching the requested output text
else:
    print("Not found")
```

# Calculate the mean of a list of marks.

```python
marks = [70, 80, 65, 75]

# Sum divided by the number of elements
average = sum(marks) / len(marks)

print(f"Average Marks: {average}")
```
