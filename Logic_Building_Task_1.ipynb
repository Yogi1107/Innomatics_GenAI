```python
# 1. User Login Check
username_input = 'admin'
password_input = '1234'

username = 'admin'
password = '1234'

if username_input == username and password_input == password:
    print('Login Successful')
else:
    print('Invalid Credentials')
```

```python
# 2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print('Total Pass Students:', pass_count)
print('Total Fail Students:', fail_count)
```

```python
# 3. Simple Data Cleaner
names = [' Alice ', 'bob', ' CHARLIE ']

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print('Cleaned Names:', cleaned_names)
```

```python
# 4. Message Length Analyzer
messages = ['Hi', 'Welcome to the platform', 'OK']

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")
    
    if length > 10:
        print('⚠️ This message is longer than 10 characters')
```

```python
# 5. Error Message Detector
logs = ['INFO', 'ERROR', 'WARNING', 'ERROR']

error_count = logs.count('ERROR')

print('Total ERROR entries:', error_count)
```
